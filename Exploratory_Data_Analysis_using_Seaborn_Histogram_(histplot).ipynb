```python
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.dropna(inplace=True)
penguins.shape
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.flipper_length_mm);
```

```python
sns.histplot(x='flipper_length_mm', data=penguins);
```

```python
sns.histplot(y = 'flipper_length_mm', data=penguins);
```

```python
sns.histplot(x = 'flipper_length_mm', data=penguins, kde=True);
```

```python
sns.histplot(x = 'flipper_length_mm', data=penguins, bins=20);
```

```python
sns.histplot(x = 'flipper_length_mm', data=penguins, bins=[170,190,200,205,215,230]);
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, binwidth = 10);
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, binwidth = 10, binrange=(170,230));
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, stat='count');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, stat='density');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, stat='probability');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue='sex');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue='sex', element='step');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue='sex', element='poly');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue='sex', multiple='layer');
```

```python
sns.histplot(x = 'flipper_length_mm', data = penguins, hue='sex', multiple='stack');
```

```python
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm', data = penguins);
```

```python
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm', cbar=True, data = penguins);
```

```python
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm', cbar=True, hue='sex', data = penguins);
```

```python
sns.histplot(x = 'species', hue = 'sex', multiple='dodge',data = penguins);
```

```python
sns.histplot(x = 'species', hue = 'sex', multiple='dodge',shrink=0.8, data = penguins);
```

```python
sns.histplot(x = 'flipper_length_mm', color='indigo', data = penguins);
```

```python
sns.histplot(x = 'species', hue = 'sex', multiple='dodge',shrink=0.8, palette='bone', data = penguins);
```

```python
sns.histplot(x = 'flipper_length_mm', color='indigo', fill=False, data = penguins);
```
