```python
## my_taipy_app.py

from taipy import Gui
```

```python
## my_taipy_app.py

data = {"Rating":[1,4,3,1,3,5,2,4,2],
        "Salary":[100,450,340,120,290,600,250,430,220]}

page = """
# First Taipy Web Application.

DataFrame: <|{data}|table|>
"""
```

```python
## my_taipy_app.py

Gui(page=page).run()
```
