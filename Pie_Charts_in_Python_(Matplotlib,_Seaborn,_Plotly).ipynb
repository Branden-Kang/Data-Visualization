# Matplotlib

```python
import matplotlib.pyplot as plt
shares = [31,20,11,7,5,4,22]
labels = ["Good Year", "Michelin", "Firestone",
          "Bridgestone", "Falken", "Continental",
          "Other"]
fig = plt.figure(figsize = (8, 8))
plt.pie(shares,labels=labels)
plt.title('Market Shares of Tire Brands in US')
plt.show()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
colors = sns.color_palette('pastel')[0:len(labels)]
plt.pie(shares, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

# Plotly

```python
import plotly.express as px
fig = px.pie(values=shares, names=labels)
fig.update_layout(title="Market Shares of Tire Brands in US")
fig.update_traces(textposition='inside',
                  textinfo='percent+label', showlegend=False)
```

## Donut Charts

```python
import plotly.graph_objs as go

outer_values = [35,25,10,25,5]
inner_values = [5,30,35,10,20]
outer_labels = ["A","B","C","D","E"]
inner_labels = ["a","b","c","d","e"]

colors = ['gold', 'mediumturquoise']*int(len(outer_values))

#inner circle
trace1 = go.Pie(
    hole=0.5,
    sort=False,
    direction='clockwise',
    domain={'x': [0.15, 0.85], 'y': [0.15, 0.85]},
    values=inner_values,
    labels=inner_labels,
    showlegend=False,
    textposition='inside',
    marker={'line': {'color': 'white', 'width': 1}, 'colors': colors}
)

#outer circle
trace2 = go.Pie(
    hole=0.7,
    sort=False,
    direction='clockwise',
    values=outer_values,
    labels=outer_labels,
    textinfo='label',
    textposition='outside',
    showlegend=False,
    marker={'colors': ['green', 'red', 'blue'],
            'line': {'color': 'white', 'width': 1}}
)

#figure
fig = go.FigureWidget(data=[trace1, trace2])
fig.layout.title = 'Nested Donut Chart'
fig.show()
```
