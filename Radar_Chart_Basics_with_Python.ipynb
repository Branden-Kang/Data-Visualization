```
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
```

```
plt.polar()
```

```
values = [5,5,6,7,5,2]
N = len(values)

values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]
```

```
plt.polar(angles,values)
```

```
categories = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
N = len(categories)

values = [18, 13, 20, 7, 9, 7]
values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]
```

```
plt.polar(angles, values)
plt.xticks(angles[:-1],categories)
plt.show()
```

```
categories = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
N = len(categories)

values = [18, 13, 20, 7, 9, 7]
values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values)
plt.xticks(angles[:-1],categories)

# ax.set_rlabel_position(0)
plt.yticks([5,10,15], color = 'grey', size = 10)
plt.ylim(0,20)

plt.show()
```

```
categories = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
N = len(categories)

values = [18, 13, 20, 7, 9, 7]
values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values)
plt.fill(angles, values, alpha =0.3)
plt.xticks(angles[:-1],categories)

# ax.set_rlabel_position(0)
plt.yticks([5,10,15], color = 'grey', size = 10)
plt.ylim(0,20)

plt.show()
```

```
categories = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
N = len(categories)

values = [18, 13, 20, 7, 9, 7]
values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, marker = '.')
plt.fill(angles, values, alpha =0.3)
plt.xticks(angles[:-1],categories)

# ax.set_rlabel_position(0)
plt.yticks([5,10,15], color = 'grey', size = 10)
plt.ylim(0,20)

plt.show()
```

```
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(121)

plt.plot()

ax = plt.subplot(122, polar="True")

categories = ['Str', 'Dex', 'Con', 'Int', 'Wis', 'Cha']
N = len(categories)

values = [18, 13, 20, 7, 9, 7]
values += values[:1]

angles = [n/float(N)*2*pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, marker = '.')
plt.fill(angles, values, alpha =0.3)
plt.xticks(angles[:-1],categories)

ax.set_rlabel_position(0)
plt.yticks([5,10,15], color = 'grey', size = 10)
plt.ylim(0,20)

plt.show()
```
