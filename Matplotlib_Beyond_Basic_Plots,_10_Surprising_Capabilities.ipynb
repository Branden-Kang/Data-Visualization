# 1 Streamplot

```python
import numpy as np
import matplotlib.pyplot as plt

Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V)
plt.show()
```

# 2 Logit Scale

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogitLocator, LogitFormatter
from matplotlib.scale import LogitScale

x = np.linspace(-6, 6, 1000)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y)
plt.yscale("logit")
plt.gca().yaxis.set_major_locator(LogitLocator())
plt.gca().yaxis.set_major_formatter(LogitFormatter())
plt.show()
```

# 3 Table

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(5, 3)
columns = ('A', 'B', 'C')
rows = ['Row %d' % i for i in range(1, 6)]

fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
ax.table(cellText=data, colLabels=columns, rowLabels=rows, loc='center')
plt.show()
```

# 4 Tricontour and Tricontourf

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)
z = x**2 + y**2

plt.tricontour(x, y, z)
plt.tricontourf(x, y, z)
plt.savefig('trico.png', dpi=150)
plt.show()
```

# 5 Broken Barh

```python
import matplotlib.pyplot as plt

data = [(10, 20), (50, 25)]
plt.broken_barh(data, (5, 10), facecolors=('orange', 'blue'))
plt.show()
```

# 6 Quiver

```python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))
u = np.cos(x) * y
v = np.sin(y) * x

plt.quiver(x, y, u, v)
plt.show()
```

# 7 Hexbin

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar()
plt.show()
```

# 8 Eventplot

```python
import numpy as np
import matplotlib.pyplot as plt

data = [np.random.rand(10), np.random.rand(15), np.random.rand(12)]

plt.eventplot(data, colors=['red', 'green', 'blue'])
plt.show()
```

# 9 Specgram

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 1/8000)
y = np.sin(2*np.pi*440*t)

plt.specgram(y, NFFT=256, Fs=8000, noverlap=128, cmap='inferno')
plt.colorbar()
plt.show()
```

# 10 Text Annotations with Arrows

```python
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [0, 1, 2]

plt.plot(x, y, 'ro-')

plt.annotate('First Point (0, 0)', xy=(0, 0), xytext=(0.5, 0.5),
             arrowprops=dict(arrowstyle='->', lw=1.5),
             fontsize=12)

plt.annotate('Second Point (1, 1)', xy=(1, 1), xytext=(1.5, 0.5),
             arrowprops=dict(arrowstyle='->', lw=1.5),
             fontsize=12)

plt.annotate('Third Point (2, 2)', xy=(2, 2), xytext=(1.5, 1.5),
             arrowprops=dict(arrowstyle='->', lw=1.5),
             fontsize=12)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Text Annotations with Arrows')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.sin(2 * x)
y3 = np.sin(3 * x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', linewidth=2, color='blue')
plt.plot(x, y2, label='sin(2x)', linewidth=2, color='green')
plt.plot(x, y3, label='sin(3x)', linewidth=2, color='red')

plt.annotate('sin(x)', xy=(5.5, 0.7), fontsize=12, color='blue')
plt.annotate('sin(2x)', xy=(5.5, -0.3), fontsize=12, color='green')
plt.annotate('sin(3x)', xy=(5.5, -0.8), fontsize=12, color='red')

plt.title('Colorful Sine Waves', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)

plt.show()
```
