```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
temperature_df = pd.read_csv('/kaggle/input/historical-hourly-weather-data/temperature.csv')
temperature_df.head()
```

```python
temperature_df.tail()
```

```python
modified_df = temperature_df.iloc[::24]
modified_df.head()
```

```python
nyc_df = temperature_df[['datetime','New York']]
nyc_df = nyc_df.iloc[::24]
nyc_df.dropna(how='any', inplace=True)
print(nyc_df.head(5))
print(nyc_df.tail(5))
```

(x − 273.15) × 9/5 + 32

```python
nyc_df['temp'] = nyc_df['New York'].apply(lambda x: (x-273.15) * 9/5 + 32)
```

```python
nyc_df.rename(columns = {'datetime':'date'},inplace=True)
```

```python
nyc_df.drop('New York',axis=1,inplace=True)
```

```python
nyc_df.head()
```

```python
nyc_df.set_index('date',inplace=True)
```

```python
print(nyc_df.head(5))
```

```python
nyc_df.reset_index(inplace=True)
```

```python
print(nyc_df.info())
```

```python
nyc_df['date'] = pd.to_datetime(nyc_df['date'])
```

```python
nyc_df['year'] = nyc_df['date'].dt.year
```

```python
nyc_df['day'] = nyc_df['date'].dt.dayofyear
```

```python
nyc_df.head()
```

```python
plt.figure(figsize=(15, 7))
nyc_chart = sns.lineplot(x="day", y="temp", hue='year', data=nyc_df ).set_title('New York Weather Over Time')
plt.show()
```
