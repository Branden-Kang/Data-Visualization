```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle
from matplotlib.patheffects import withStroke
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

# Ignore warning
import warnings
warnings.filterwarnings("ignore")

# Define Figure，it is within the red frame in the picture
fig = plt.figure(figsize=(9, 8), facecolor='1', dpi=150)

# Figure title
fig.suptitle('Matplotlib Details', x=0.46, fontsize=20, ha='right')

# Add axes on the figure
marg = 0.15
ax = fig.add_axes([marg, marg, 1 - 1.8 * marg, 1 - 1.8 * marg],
                  aspect=1,
                  facecolor='0.9')

# Data
np.random.seed(19680801)
X = np.linspace(0.5, 3.5, 120)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))

# Draw on the axes
ax.plot(X, Y1, c='orange', lw=1, label="Orange signal", zorder=10)
ax.plot(X[::3],
        Y3[::3],
        linewidth=0,
        markersize=6,
        marker='*',
        markerfacecolor='none',
        markeredgecolor='black',
        markeredgewidth=1)

# Axes title
ax.set_title("Matplotlib Elements", fontsize=15, verticalalignment='bottom')

# Legend
ax.legend(loc="upper right", fontsize=10)

# Set x and y labels
ax.set_xlabel("x Label", fontsize=12)
ax.set_ylabel("y Label", fontsize=12)

# Set x and y locators
ax.xaxis.set_major_locator(MultipleLocator(1.000))  # Major locator of x
ax.xaxis.set_minor_locator(AutoMinorLocator(4))  # Minor locator of x

ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))


# Format locator
def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return f"{x:.2f}"


ax.xaxis.set_minor_formatter(minor_tick)

# Locator range
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# Locator font, size
ax.tick_params(which='major', width=1.0, labelsize=12)
ax.tick_params(which='major', length=10, labelsize=12)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=6, labelcolor='0.5')

# Gridlines
ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

# Annotation
ax.annotate(
    "",
    xy=(4, 4),
    xytext=(4.2, 2.2),
    color=(0.25, 0.25, 1.00),
    weight="regular",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="black"),
)

ax.annotate(
    "",
    xy=(4, 0),
    xytext=(4.2, 1.8),
    color=(0.25, 0.25, 1.00),
    weight="regular",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="black"),
)

# Rectangle edge
fig.add_artist(
    Rectangle((0, 0),
              width=1,
              height=1,
              facecolor='none',
              edgecolor='red',
              linewidth=1.0))


# Circle note
def just_circle(x, y, radius=0.15):
    c = Circle((x, y),
               radius,
               clip_on=False,
               zorder=10,
               linewidth=0.6,
               edgecolor='black',
               facecolor='none',
               path_effects=[withStroke(linewidth=5, foreground=(1, 1, 1, 1))])
    ax.add_artist(c)


# Text note
def text(x, y, text):
    ax.text(x,
            y,
            text,
            zorder=100,
            ha='center',
            va='top',
            weight='bold',
            color='black',
            style='italic',
            path_effects=[withStroke(linewidth=7, foreground=(1, 1, 1, 1))])


# matplotlib methods note
def code(x, y, text):
    ax.text(x,
            y,
            text,
            zorder=100,
            ha='center',
            va='top',
            weight='normal',
            color=(0.25, 0.25, 1.00),
            fontsize='medium',
            path_effects=[withStroke(linewidth=7, foreground=(1, 1, 1, 1))])


def circle(x, y, txt, cde, radius=0.1):
    just_circle(x, y, radius=radius)
    text(x, y - 0.2, txt)
    code(x, y - 0.33, cde)


circle(4.385, 4.3, "Figure", "plt.figure")
circle(4.3, 2.2, "Subplot, the entire shadow part", "fig.subplots")

circle(-0.67, 4.43, "Figure title", "fig.suptitle")
circle(1.08, 4.13, "Subplot title", "ax.set_title")

circle(1.75, 2.80, "Line chart", "ax.plot")
circle(1.5, 1.64, "Marker shape", "ax.plot,marker")
circle(3.00, 3.00, "Gridlines", "ax.grid")
circle(2.8, 3.65, "Legend", "ax.legend")

circle(-0.03, 1.05, "Major locator", "ax.yaxis.set_major_locator")
circle(-0.15, 3.00, "Major locator label", "ax.yaxis.set_major_formatter")
circle(0.00, 3.75, "Minor locator", "ax.yaxis.set_minor_locator")
circle(3.25, -0.10, "Minor locator label", "ax.xaxis.set_minor_formatter")

circle(0.65, 0.01, "x axis", "ax.xaxis")
circle(0, 0.44, "y axis", "ax.yaxis")
circle(1.650, -0.32, "x axis title", "ax.set_xlabel")
circle(-0.47, 1.68, "y axis title", "ax.set_ylabel")

circle(4.0, 0.7, "Spines", "ax.spines")
circle(-1.17, -0.22, "Edge", "fig.add_artist")

circle(4.1, 3.5, "Annotation", "ax.annotate/text")

plt.show()
```
