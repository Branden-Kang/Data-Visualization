```python
fake_country = {
    '0-4 years'   : [169, 48, 52],
    '5-9 years'   : [191, 46, 54],
    '10-14 years' : [211, 47, 53],
    '15-19 years' : [212, 44, 56],
    '20-24 years' : [222, 32, 68],
    '25-34 years' : [438, 35, 65],
    '35-44 years' : [515, 40, 60],
    '45-54 years' : [480, 42, 58],
    '55-64 years' : [338, 39, 61],
    '65-74 years' : [221, 38, 62],
    '75+ years'   : [192, 39, 61], }
plt.pie(fake_country['0-4 years'][1:])
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(16,8))
values = fake_country['0-4 years']
plt.pie(values[1:], autopct='%.0f%%', 
        radius = math.sqrt( (values[0]/100)/ math.pi), 
        textprops={'fontsize':values[0]/10})
plt.show()
```

```python
previous_r, center = 0, 0
centers = []
for i, key in enumerate(fake_country.keys()):
    # get array with values
    values = fake_country[key]
    # calculate radius
    r = (math.sqrt( (values[0])/ math.pi) *-1)
    # get center and append to list
    center += previous_r + r
    centers.append(center)
    # replace previous radius
    previous_r = -5 + r   
    
centers
```

```python
fig, ax = plt.subplots(1, figsize=(12,24))
for i, key in enumerate(fake_country.keys()):
    values = fake_country[key]
    plt.pie(values[1:], 
            autopct='%.0f%%', 
            radius=math.sqrt( (values[0])/ math.pi), 
            textprops={'fontsize':12}, 
            center=(0,centers[i]), 
            colors=['#2482B6', '#F34D3A'], 
            frame= True)
plt.xlim(-10,10)
plt.ylim(-250, 1)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12,24))
for i, key in enumerate(fake_country.keys()):
    values = fake_country[key]
    labels = [str(val)+'%' for val in values[1:]]
    plt.pie(values[1:], autopct='%.0f%%', radius=math.sqrt( (values[0])/ math.pi), 
            textprops={'fontsize':12}, center=(0,centers[i]), 
            colors=['#2482B6', '#F34D3A'])
    
    plt.text(15,centers[i], key, fontsize=14)
plt.xlim(-10,10)
plt.ylim(-250, 1)
plt.title('Age and Sex Proportions of the Population\n\n\n', loc='left', fontsize=14)
plt.legend(['Male', 'Female'], ncol=2, 
          bbox_to_anchor=(1.4, 1.03))
plt.show()
```
