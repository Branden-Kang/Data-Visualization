# Conversion of University Rating


```python
print(df.info())
```

```python
df['University Rating'] = df['University Rating'].astype(str)
```

# Plotting the Ridgeline Plot


```python
joyplot(df, by = 'University Rating', column = 'CGPA') 
plt.xlabel("CGPA")
plt.show()
```

# Beautifying the Ridgelines Plot


## 1. Customize Plot Colours and Fade


```python
joyplot(df, by = 'University Rating', column = 'CGPA', color = 'Orange', fade = True) 
plt.show()
```

```python
from matplotlib import cm 
joyplot(df, by = 'University Rating', column = 'CGPA', colormap=cm.autumn, fade = True) 
plt.show()
```

## 2. Customize Plot Layout


```python
joyplot(df, by = 'University Rating', column = 'CGPA', colormap = cm.autumn, fade = True, range_style='own', figsize = (10,6)) 
plt.show()
```

## 3. Adding title to Ridgeline Plot


```python
joyplot(df, by = 'University Rating', column = 'CGPA', colormap = cm.autumn, fade = True, range_style='own', figsize = (10,6), title = 'Distribution of Student CGPA based on University Rating') 
plt.show()
```

## 4. Plot Histogram instead of Density Curve


```python
joyplot(df, by = 'University Rating', column = 'CGPA', color = 'Orange', fade = True, range_style='own', figsize = (10,6), hist = True, overlap = 0, title = 'Distribution of Student CGPA based on University Rating') 
plt.show()
```
