```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Import Datasets
exercise = sns.load_dataset('exercise')
iris = sns.load_dataset('iris')
penguins = sns.load_dataset('penguins')
mpg = sns.load_dataset('mpg')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
```

# 1. Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a bar plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill by Day")
plt.show()
```

# 2. Count Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the mpg dataset
mpg = load_dataset("mpg")

sns.countplot(x="class", data=titanic)
plt.title("Passenger Count by Class")
plt.show()
```

# 3. Histograms

```python
sns.histplot(mpg["mpg"], bins=20)
plt.title("Distribution of Miles Per Gallon (mpg)")
plt.show()
```

# 4. Cat Plots (Box, Violin, Swarm, Boxen)

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

# Load the iris dataset
iris = load_dataset("iris")

sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

# 5. Multiple Plots using FacetGrid

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

# 6. Joint Plots

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.title("Joint Plot of Sepal Length and Sepal Width")
plt.show()
```

# 7. KDE Plots

```python
sns.kdeplot(penguins["flipper_length_mm"])
plt.title("KDE Plot of Flipper Length in Penguins")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import load_dataset

titanic = sns.load_dataset("titanic")

# Select only numeric columns for the heatmap
numeric_titanic = titanic.select_dtypes(include=["float64", "int64"])

sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
