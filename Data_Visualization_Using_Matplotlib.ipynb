```python
# Importing Matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl

# For plotting in jupyter notebook
%matplotlib inline 

# Import more packages to generate data
import numpy as np
```

```python
# Fake up data for our plots
x1 = np.arange(-3 * np.pi, 3 * np.pi, np.pi/6)
x2 = np.arange(-3 * np.pi, 3 * np.pi, np.pi/3)

y1 = np.sin(x1)
y2 = np.cos(x2)

plt.plot(x1, # Values on x-axis
         y1, # Values on y-axis
         label='sin' # Label of the plot
         )

# Add title
plt.title('Simple plot')

# Add label to axis
plt.xlabel('X')
plt.ylabel('Y')

# Add grids
plt.grid()

# Add another plot
plt.plot(x2, y2, label='cos')

# Add legend
plt.legend()
```

```python
# Let's add more parameters to our plot
plt.plot(x1, y1,
         label='sin',   # Label of the plot
         color='teal',  # Color of the line
         marker='v'     # Shape of the marker
         )    

# Add title
plt.title('Simple plot')

# Add label
plt.xlabel('X')
plt.ylabel('Y')

# Add grids
plt.grid()

# Add another plot
plt.plot(x2, y2, label='cos', color='plum', marker='o',
         linestyle='--',    # Line style
         linewidth=2        # Width of the line
         )

# Add legend
plt.legend(['Cosine', 'Sine'], # Labels
           loc='upper left'     # Location of the legend
           )    

# Change the y-ticks
plt.yticks(np.arange(-1, 1, 0.5))
```

```python
# Let's add a text to a plot with specific color and make it bold and italic

arrowprops = dict(arrowstyle="->",  
                  connectionstyle="arc3,rad=.3", # Specify the connection style
                  color='red',  # Set the color
                  lw=1  # Set the width of line
                  )

plt.plot(x2, y2)
plt.annotate('Local max', # text to annotate
             xy=(0., 1), # the position of x and y to point
             xytext=(1.5,.5), # the position of text
             color='brown', # color of text
             arrowprops=arrowprops, # property of the arrow
             fontsize=13, fontweight='bold', fontstyle='italic')


plt.annotate('Local min', # text to annotate
             xy=(-3.1,-1), 
             xytext=(-5, -.2), # the xy position to add the texts
             bbox=dict(boxstyle='round', fc='yellow'), # or property of text box
             arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3));
```

```python
# Let's plot a heart 

theta = np.linspace(0, 2 * np.pi, 100)
x = 16 * ( np.sin(theta) ** 3 )
y = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)

# font
hfont = {'fontname':'Apple Chancery',
         'fontsize': 14,
         'color': 'w'}

# Plotting
plt.plot(x, y)

# Annotation with defining font 
plt.annotate('Heart', xy=(-1.5 ,0), **hfont)

# Filling between two lines
plt.fill_between(x, y)

# Turning off the axis
plt.axis('off');
```

```python
# Let's define a font style for title of plots first 
title_fontdic= {'size':9, 'color':'red', 'fontname':'Segoe Print', 'family':'serif'}


# Let's create a figure object

fig = plt.figure(figsize=(5, 2), 
           dpi=100, 
           facecolor='yellow' 
           )

ax1 = fig.add_subplot(1, 2, 1) # adding a subplot to figure object
ax1.plot(x1, y1)  # plot using ax object

ax1.set_facecolor('#E2D784') # set the face color for axis

ax1.set_xlabel('X')
ax1.set_ylabel('Y', labelpad=-10) # Decreasing the pad of label
ax1.set_title('Plot 1 \U0001F60A', loc='left', fontsize=10)
ax2 = fig.add_subplot(1, 2, 2) # adding another subplot to figure object
ax2.plot(x2, y2, 'lime')  # plot using ax object

# Set the face color for axis
ax2.set_facecolor('#F68989') 

# Set xlabel
ax2.set_xlabel('X')

# Deleting the y-ticks for second matrix
ax2.set_yticks([])  

# Set title for second plot
ax2.set_title('plot 2', loc='right', fontdict=title_fontdic) 

plt.subplots_adjust(wspace=0.1)  # reduce the space between two axis
```

```python
# Let's  generate some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
# A simple plot

fig = plt.figure(figsize=(6, 3))
ax= fig.add_subplot(121) 
ax.boxplot(data)
ax.set_title('Box plot')


ax= fig.add_subplot(122) 
ax.set_title('Hist plot')
ax.hist(data);
```

```python
fig = plt.figure(figsize=(3, 3))
ax= fig.add_subplot(111) 
ax.boxplot(data, 
           notch=True, # To put a notch on median
           sym='x', # Marker for outliers
           vert=False, # Plot horizontal
           patch_artist=True,
           capprops=dict(  # Property of caps
               color = '#062C30',  # Color
               lw = 2.5, # Line width
           ),
           whiskerprops = dict( # Property of whisker
               color = '#39AEA9',   # Color
               lw = 1.5,  # Line width
               ),
            medianprops = dict(  # Property of median line
                color = 'blue',  # Color
                lw = 1.5    # Line width
                ),
            boxprops = dict( # Properties of the box
                facecolor = '#F1E1A6', # Background color
                edgecolor = '#062C30',  # COlor line
                lw = 1.5  # Line width
                ),
           );
```

```python
# A simple plot

fig = plt.figure(figsize=(6, 3))
ax= fig.add_subplot(121) 
ax.violinplot(data)
ax.set_title('Violin plot')


ax= fig.add_subplot(122) 
ax.set_title('Box plot')
ax.boxplot(data);
```

```python
from matplotlib.animation import FuncAnimation

# Enable intractive plot
%matplotlib notebook

fig = plt.figure()
ax = plt.axes(xlim=(0, 5), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)


def init():
    line.set_data([], [])
    return line, 

    
def animate(i):
    x = np.linspace(0, 5, 1000)
    f = 1
    y = np.sin(2 * f * np.pi * x + i)
    line.set_data(x, y)
    return line,

    
anim = FuncAnimation(fig, animate, init_func=init,
                     frames=200, interval=40, blit=True)
plt.show()
```

```python
fig = plt.figure()
#creating a subplot 
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    data = open('./stock.txt','r').read()
    lines = data.split('\n')
    xs = []
    ys = []
   
    for line in lines:
        x, y = line.split(',') # Delimiter is comma    
        xs.append(float(x))
        ys.append(float(y))
   
    
    ax1.clear()
    ax1.plot(xs, ys)

    plt.xlabel('Day of trade')
    plt.ylabel('Price')
    plt.title('Live graph with Matplotlib')

    
ani = FuncAnimation(fig, animate, interval=1000) 
plt.show()
```
