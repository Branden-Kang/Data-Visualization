```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
tips_data = sns.load_dataset('tips')

# Create a scatter plot with hue for categorical data
sns.scatterplot(data=tips_data, x='total_bill', y='tip', hue='sex')

# Set the title and labels for the graph
plt.title('Total Bill vs. Tip Amount')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Load the dataset
air_passengers = sm.datasets.get_rdataset("AirPassengers", "datasets").data

# Plot the original data with every single data point
plt.figure(figsize=(10, 6))
plt.plot(air_passengers.index, air_passengers["value"])
plt.title("Monthly International Airline Passengers (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()

# Plot the same data with a rolling average
rolling_avg = air_passengers["value"].rolling(window=12).mean()
plt.figure(figsize=(10, 6))
plt.plot(air_passengers.index, rolling_avg)
plt.title("Monthly International Airline Passengers (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()
```
