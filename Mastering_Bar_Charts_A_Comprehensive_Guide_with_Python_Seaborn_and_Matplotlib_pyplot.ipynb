# 1. Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Values': np.random.randint(10, 100, 100)
})

# Create a basic bar chart
sns.barplot(x='Category', y='Values', data=data, ci=None)
plt.title('Basic Bar Chart - Total Values by Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Values': np.random.randint(10, 100, 100)
})

# Calculate total values for each category
category_totals = data.groupby('Category')['Values'].sum().reset_index()

# Create a basic bar chart using Matplotlib
plt.bar(category_totals['Category'], category_totals['Values'])
plt.title('Basic Bar Chart - Total Values by Category')
plt.xlabel('Category')
plt.ylabel('Total Values')
plt.show()
```

# 2. Clustered Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
products = ['Product 1', 'Product 2', 'Product 3']
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Product': np.random.choice(products, 100),
    'Sales': np.random.randint(10, 100, 100)
})

# Create a clustered bar chart
sns.barplot(x='Category', y='Sales', hue='Product', data=data, ci=None)
plt.title('Clustered Bar Chart - Sales by Product and Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
products = ['Product 1', 'Product 2', 'Product 3']
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Product': np.random.choice(products, 100),
    'Sales': np.random.randint(10, 100, 100)
})

# Prepare data for plotting
pivot_data = data.pivot_table(index='Category', columns='Product', values='Sales', aggfunc='sum')

# Create a clustered bar chart using Matplotlib
width = 0.2  # Width of each bar
x = np.arange(len(categories))

fig, ax = plt.subplots()
for i, product in enumerate(products):
    ax.bar(x + i * width, pivot_data[product], width=width, label=product)

ax.set_xticks(x + width * (len(products) - 1) / 2)
ax.set_xticklabels(categories)
ax.set_xlabel('Category')
ax.set_ylabel('Sales')
ax.set_title('Clustered Bar Chart - Sales by Product and Category')
ax.legend(title='Product')

plt.show()
```

# 3. Stacked Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C']
products = ['Product 1', 'Product 2', 'Product 3']
data = pd.DataFrame({
    'Category': np.random.choice(categories, 100),
    'Product': np.random.choice(products, 100),
    'Sales': np.random.randint(10, 100, 100)
})

# Prepare data for plotting
pivot_data = data.pivot_table(index='Category', columns='Product', values='Sales', aggfunc='sum')

# Create a stacked bar chart using Matplotlib
width = 0.8  # Width of each bar
bottoms = np.zeros(len(categories))

fig, ax = plt.subplots()
for product in products:
    ax.bar(pivot_data.index, pivot_data[product], width=width, label=product, bottom=bottoms)
    bottoms += np.array(pivot_data[product])

ax.set_xlabel('Category')
ax.set_ylabel('Total Sales')
ax.set_title('Stacked Bar Chart - Total Sales by Category, Stacked by Product')
ax.legend(title='Product', loc='upper right', bbox_to_anchor=(1.15, 1))

plt.show()
```
