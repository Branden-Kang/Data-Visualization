```python
import matplotlib.pyplot as plt

portfolio = {'Cash':90000,'Real Estate':700000,'Stocks':375000,
             'Watches':55000,'Baseball Cards':230000}

sorted_portfolio = sorted(portfolio.items(), 
                          key= lambda key_value: key_value[1],
                          reverse=True)

labels = [key_value_tuple[0] for key_value_tuple in sorted_portfolio]
data = [key_value_tuple[1] for key_value_tuple in sorted_portfolio]

plt.pie(data,labels=labels,startangle=90,counterclock=False)
plt.show()
```

```python
total = sum(data)
colors = ["#00ebc1", "#00b8ba", "#0085b4", "#0051ad", "#001ea6"]
percentages = ["{0:.1%}".format(value / total) for value in data]
plt.pie(data,labels=percentages,startangle=90,counterclock=False,colors=colors)
```

```python
plt.pie(data,labels=percentages,startangle=90,counterclock=False,
    colors=colors,textprops={"ha":"center"},labeldistance=1.15)
```

```python
import matplotlib.pyplot as plt

portfolio = {'Cash':90000,'Real Estate':700000,'Stocks':375000,
             'Watches':55000,'Baseball Cards':230000}

sorted_portfolio = sorted(portfolio.items(), 
                          key= lambda key_value: key_value[1],
                          reverse=True)

colors = ["#00ebc1", "#00b8ba", "#0085b4", "#0051ad", "#001ea6"]

labels = [key_value_tuple[0] for key_value_tuple in sorted_portfolio]
data = [key_value_tuple[1] for key_value_tuple in sorted_portfolio]
labels = ['{}\n${:,.0f}'.format(label,value) 
          for label, value in zip(labels,data)]

total = sum(data)
percentages = ["{0:.1%}".format(value / total) for value in data]

plt.pie(data,labels=percentages,startangle=90,counterclock=False,
    colors=colors,textprops={"ha":"center"},labeldistance=1.15)
plt.legend(labels=labels,loc=(1.02,0.25))
plt.tight_layout(pad=3)
plt.title('${:,.0f} Investments'.format(total))

plt.show()
```
