```python
import networkx as nx

G = nx.Graph()
```

```python
G.add_node('A')
G.add_node('B')
G.add_node('C')
G.add_node('D')
G.add_node('E')
G.add_node('F')
G.add_node('G')
```

```python
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G'])
```

```python
G.add_edge('A', 'B')
G.add_edge('A', 'G')
G.add_edge('A', 'D')
G.add_edge('B', 'G')
G.add_edge('B', 'C')
G.add_edge('B', 'E')
G.add_edge('C', 'E')
G.add_edge('C', 'F')
G.add_edge('D', 'F')
G.add_edge('E', 'F')
G.add_edge('F', 'G')
```

```python
G.add_edges_from([('A', 'B'), ('A', 'G'), ('A', 'D'), ('B', 'G'),  ('B', 'C'), ('B', 'E'), ('C', 'E'), ('C', 'F'), ('D', 'F'), ('E', 'F'), ('F', 'G')])
```

```python
Adjacency_list = {'A': ['B', 'G', 'D'],'B': ['A', 'G', 'C', 'E'],'C': ['B', 'E', 'F'],'D': ['A', 'F'],'E': ['B', 'C', 'F'],'F': ['G', 'D', 'C', 'E'],'G': ['A', 'B', 'F']}
G = nx.Graph(Adjacency_list)
```

```python
G.edges()
```

```python
nx.draw(G)
```

```python
nx.draw(G, pos = None, ax = None, with_labels = True,font_size = 20, node_size = 2000, node_color = 'lightgreen')
```

```python
pos =nx.spring_layout(G, seed = 50)
nx.draw(G, pos, ax = None, with_labels = True,font_size = 20, node_size = 2000, node_color = 'lightgreen')
```

```python
Adjacency_list = {'A': ['B', 'D'], 'B': ['G', 'C', 'E'], 'C': ['F'],'D': ['A', 'F'], 'E': ['C', 'F'], 'F': [], 'G': ['A', 'F']}
G = nx.DiGraph(Adjacency_list)
pos =nx.spring_layout(G, seed = 50)

nx.draw(G, pos, ax = None, with_labels = True,font_size = 20, node_size = 2000, node_color = 'lightgreen')
```
