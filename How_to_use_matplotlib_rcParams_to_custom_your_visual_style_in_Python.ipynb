````python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(dpi=100)
plt.plot([1,2,3])
````

````python
#lines.linestyle: default '-'
#lines.marker: default no marker
#axes.grid: default False

import matplotlib as mpl

mpl.rcParams['lines.marker'] = '^'
mpl.rcParams['lines.linestyle'] = '--'
mpl.rcParams['axes.grid'] = True

plt.figure(dpi=100)
plt.plot([1,2,3])

print(mpl.rcParams['lines.marker'])
````

````python
mpl.rcParams['lines.marker'] = '^'
mpl.rcParams['lines.linestyle'] = '--'
mpl.rcParams['axes.grid'] = True
mpl.rc_file_defaults()

plt.figure(dpi=100)
plt.plot([1,2,3])
````

# Use “with” to partly change style

````python
import numpy as np
import matplotlib.pyplot as plt
mpl.rc_file_defaults()
my_dpi=96
plt.figure(figsize=(480/my_dpi,480/my_dpi),dpi=my_dpi)
plt.figure()


with plt.style.context('ggplot'): # Use "with" to change the style for this subplot only
    plt.subplot(221)
    plt.plot([-1,0,1])


plt.subplot(222)# This subplot is set to default style, because the above subplot is a partly setup
plt.plot([1,2,3])

plt.style.use('fivethirtyeight') # Set up the style
plt.subplot(223)
plt.plot([1,2,3])

plt.subplot(224) # This subplot will be "fivethirtyeight" style because the last one did a global setup
plt.plot([1,2,3])
````

# Combined style

````python
mpl.rc_file_defaults()
plt.style.use('fivethirtyeight')
plt.subplot(131)
plt.plot([1,2,3])
plt.title("'fivethirtyeight'")

mpl.rc_file_defaults()
plt.style.use('ggplot')
plt.subplot(132)
plt.plot([1,2,3])
plt.title("'ggplot'")


mpl.rc_file_defaults()
plt.style.use(['fivethirtyeight', 'ggplot']) # Combine
plt.subplot(133)
plt.plot([1,2,3])
plt.title("['fivethirtyeight', 'ggplot']")
````
