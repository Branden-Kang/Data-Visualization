# Area Plots


```python
import matplotlib.pyplot as plt                       
import numpy as np                       
import pandas as pd

plt.style.use("fivethirtyeight")
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list("ABCD"))
df.head()
```

```python
df['A'].plot.area()
```

```python
df.plot.area()
```

```python
df.plot.area(stacked=False)
```
