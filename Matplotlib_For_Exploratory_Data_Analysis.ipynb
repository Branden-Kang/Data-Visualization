```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
a = np.linspace(0,5,11)
b = a ** 3
print(a)
print(b)
```

# basic plotting

```python
plt.plot(a,b)
```

# Add color and line style

```python
plt.plot(a,b,'r--')
```

# Add label and title

```python
plt.plot(a,b,'b-.')
plt.xlabel('a')
plt.ylabel('b')
plt.title('Graph')
```

# Create multiple plots using the functional method

```python
plt.subplot(1,2,1)
plt.plot(a,b,'g')
plt.subplot(1,2,2)
plt.plot(b,a,'y')
```

# Basic plotting using object oriented method

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b)
axes.set_xlabel('a')
axes.set_ylabel('b')
axes.set_title('Graph')
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='green',linewidth=10,alpha=0.5,linestyle='-.')
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='purple',linewidth=3,marker='o', markersize=10, markerfacecolor='red', markeredgewidth=2, markeredgecolor='yellow')
```

```python
fig  = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='black')
axes.set_xlim([0,2])
axes.set_ylim([0,3])
```

# Create multiple plots using object-oriented method

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.4,0.5,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.5,0.3,0.3])
axes1.plot(a,b)
axes1.set_title('Bigger plot')
axes2.plot(b,a)
axes2.set_title('Smaller plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=3)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes[0].plot(a,b)
axes[0].set_title('Graph 1')
axes[1].plot(b,a)
axes[1].set_title('Graph 2')
```

# Adjust the Figure size and DPI

```python
fig = plt.figure(figsize=(4,3))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(4,3), dpi=110)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

```python
fig = plt.figure(figsize=(9,3))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5,2))
axes[0].plot(a,b)
axes[1].plot(b,a)
```

# How to deal with legends

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b,label='b v/s a')
axes.plot(b,a,label='a v/s b')
axes.legend()
```

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b,label='b v/s a')
axes.plot(b,a,label='a v/s b')
axes.legend(loc=10)
```
