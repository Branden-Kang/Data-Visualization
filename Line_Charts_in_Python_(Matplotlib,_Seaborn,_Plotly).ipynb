```python
import random
def generate_time_data():
    years = []
    vals = []
    for i in range(100):
        years.append(1900+i)
        change = random.uniform(-1, 1)
        val = 100 + (100*change)
        vals.append(val)
    return years,vals
```

```python
years,vals = generate_time_data()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (14, 8))
plt.plot(years,vals)
plt.title('Change over Time')
plt.xlabel('Years')
plt.ylabel('Val')
plt.show()
```

```python
import seaborn as sns
ax = sns.lineplot(x=years,y=vals)
ax.set(xlabel='Years', ylabel='Vals', title='Change over Time')
plt.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objs as go
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(
    go.Bar(name="bars", x=years, y=vals),
    secondary_y=False,
    )
fig.add_trace(
    go.Scatter(name="vals",x=years, y=vals,
              mode='lines', marker=dict(size=12)),
    secondary_y=True,
    )
fig.update_layout(title_text="Double Axis Chart")
fig.update_xaxes(title_text="Years")
fig.update_yaxes(title_text="Bars", secondary_y=False)
fig.update_yaxes(title_text="Vals", secondary_y=True)
```
