```python
import pandas as pd
import plotly.express as px
pd.set_option('display.max_rows',100)

df = pd.read_csv("https://raw.githubusercontent.com/jonyjesus18/visuals/master/co2/owid-co2-data.csv")
df = df[df['year'] > 1970].dropna()
df = df[df['country'].isin(['Austria','Brazil','Belgium','Canada','France','Italy','Portugal','Chile','South Africa','India'])]

import plotly.express as px
fig = px.line(df, x='year', y=['co2'],
labels='co2',color = df['country'],title = 'CO2 Emissions')
fig.show()
```

```python
import pandas as pd
import yfinance as yf
import plotly.express as px
pd.set_option('display.max_rows',None)
st = '2022-01-05' #wed
end = '2022-04-20' 
data = pd.read_csv('https://raw.githubusercontent.com/jonyjesus18/visuals/master/fin/constituents.csv')
symbol = data["Symbol"].tolist()
sector = data["Sector"].tolist()

df = pd.DataFrame()
for i in range(len(symbol)):
    removed = []
    try:
        a = yf.download([f'{symbol[i]}'], 
       start=st,end=end,progress=False,interval = '1wk'
        )
        a['Ticker'] = symbol[i]
        a['Area'] = sector[i]
        a['Rate'] = (100*a['Close']/a['Close'][0])-100
        df = pd.concat([df, a], axis=0)
        print("download number: #",i )
    except:
        removed = removed.append(symbol[i])
        
def cut(x):
    return str(x)[:-9]
    
df['date'] = df.index
df.date = df.date.apply(cut)
df = df.dropna()
df.head()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://raw.githubusercontent.com/jonyjesus18/visuals/master/fin/table.csv')

fig = px.scatter(df, x = df['Volume'], 
y = df.Rate, animation_frame=df.date, animation_group=df.Ticker,
size=df['Volume'],title = 'S&P 500 Rate of Change since the 1st Jan 2022',color=df.Area, hover_name=df.Ticker,log_x=True, size_max=40, range_x=[10000,1000000000], range_y=[min(df['Rate']),max(df['Rate'])])
           
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 150

import plotly.io as pio
pio.renderers.default = 'notebook_connected'
fig.show(renderer='notebook_connected')
```

```python
#Get data and Clean Coordinates

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/jonyjesus18/visuals/master/fireballs/nasas%20fireballs.csv')
df.head(10)

def cleanCoordinate(coordinate):
    c = coordinate[-1]
    if c =='N': coordinate = f"+{coordinate[:-1]}"
    if c =='S': coordinate = f"-{coordinate[:-1]}"
    if c =='E': coordinate = f"+{coordinate[:-1]}"
    if c =='W': coordinate = f"-{coordinate[:-1]}"
    return float(coordinate)
    
df = df.drop_duplicates()
df = df.dropna()
df['latitude'] = df['latitude'].apply(cleanCoordinate)
df['longitude'] = df['longitude'].apply(cleanCoordinate)
df['velocity(km/s)'] = abs(df['velocity(km/s)'])
```

```python
#Construct main Dataframe for plotting

df2 = pd.DataFrame()
dates = df['date/time for peak brightness'].sort_values(ascending = True).tolist()

for date in dates:
    # tdf = df[df['date/time for peak brightness'] <= date]
    tdf = df.loc[df['date/time for peak brightness'] <= date]
    tdf['date/time for peak brightness'] = date
    df2 = pd.concat([df2,tdf], axis = 0)
    
df2 = df2.sort_values(by=['date/time for peak brightness'],ascending = True)
df2 = df2.reset_index()
df2 = df2.drop(columns=['index'], axis=1)
import statistics

df2
a = statistics.mean(df['total impact energy'])
b = statistics.stdev(df['total impact energy'])

df2 = df2.drop(df2[df2['total impact energy'] > a*b].index)
```

```python
import plotly.express as px
fig = px.scatter_mapbox(df2, lat="latitude", lon="longitude", 
color="total impact energy",
range_color=[min(df2['total impact energy']),max(df2['total impact energy'])],
size="velocity(km/s)",
size_max=20,
zoom=0.75,
labels='Asteroids',
animation_frame='date/time for peak brightness'
)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(
    title='Meteor Impacts between 1998 and 2021',
    title_x=0.5,
    yaxis_title ='Total Impact Energy (MJ)'
)
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 125

import plotly.io as pio
pio.renderers.default = 'notebook_connected'

fig.show(renderer='notebook_connected')
```
