```python
from plotnine import *
import pandas as pd

# Create a sample dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5],
                   'y': [2, 4, 6, 8, 10]})

# Create a scatter plot with regression line
(ggplot(df, aes('x', 'y')) +
 geom_point() +
 geom_smooth(method='lm', se=False) +
 labs(title='Scatter plot with Regression Line',
      x='X Axis Label', y='Y Axis Label'))
```

# Step 2: Import Required Libraries

```python
import pandas as pd
from plotnine import *
```

# Step 3: Load Data

```python
from plotnine.data import mpg

# Load the mpg dataset
df = mpg
```

# Step 4: Create a Basic Plot

```python
# Create a scatter plot
scatter_plot = (
    ggplot(df, aes(x='displ', y='hwy', color='class'))
    + geom_point()
    + labs(title='Engine Displacement vs. Highway MPG', x='Engine Displacement (L)', y='Highway MPG')
)
print(scatter_plot)
```

# Step 5: Customize Plot Appearance

```python
# Customize plot appearance
scatter_plot = (
    scatter_plot
    + theme_minimal()  # Set the plot theme
    + scale_color_brewer(type='qual', palette='Set1')  # Change color palette
    + labs(title='Engine Displacement vs. Highway MPG', x='Engine Displacement (L)', y='Highway MPG')
    + theme(plot_title=element_text(size=16, face='bold'),  # Change title appearance
            axis_text=element_text(size=12),  # Change axis text appearance
            axis_title=element_text(size=14, face='italic'))  # Change axis title appearance
)
print(scatter_plot)
```

# Step 6: Add Additional Layers

```python
# Add a linear regression line to the scatter plot
scatter_plot_with_line = (
    scatter_plot
    + geom_smooth(method='lm', se=False, color='black')
)
print(scatter_plot_with_line)
```

# Step 7: Save and Display the Plot

```python
# Save the plot as a PNG image
scatter_plot_with_line.save(filename='scatter_plot.png', format='png', width=8, height=6, dpi=300)
```
