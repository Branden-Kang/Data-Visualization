```python
df_artist = pd.read_csv("/kaggle/input/spotify-dataset-19212020-160k-tracks/data_by_artist.csv")
```

```python
df_artist.head()
```

```python
top_artists = df_artist[['artists','count']].sort_values('count',ascending=False)[0:10]
```

```python
plt.figure(figsize=(10,5))
plot=sns.barplot('artists','count',data=top_artists,palette='vlag')
plot.set_xticklabels(plot.get_xticklabels(),rotation=45)
plot.set(xlabel='Artist',ylabel='Number of Songs')
```

```python
sns.jointplot(x='energy',y='danceability',data=df_artist[['energy','danceability']],
             kind='hex').set_axis_labels("Energy","Danceability")
```

```python
df_artist[['artists','danceability']].sort_values('danceability',ascending=False)[0:10]
```

```python
young_names = [i for i in df_artist['artists'] if "young" in i.lower()]
df_artist["Young"] = df_artist['artists'].apply(lambda x: 1 if x in young_names else 0)
```

```python
young_artists = df_artist.query('Young==1')
young_artists.mean() > df_artist.mean()
```
