```python
import yfinance as yf
import pandas as pd
```

```python
#Creating Ticker() Object
ticker = yf.Ticker('PLTR')
```

```python
#Creating cash flow table for Power BI
#Creating data frame & formatting
cashflow_df = ticker.cashflow.transpose().reset_index().rename(columns={'':'Year'})
#Changing data types
cashflow_df['Year']= cashflow_df['Year'].astype(str).str[:4]
#Finding free cash flow
cashflow_df['Free Cash Flow'] = cashflow_df['Total Cash From Operating Activities'] + cashflow_df['Capital Expenditures']
```
