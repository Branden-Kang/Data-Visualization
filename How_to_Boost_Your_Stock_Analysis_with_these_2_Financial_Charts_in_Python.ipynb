# 1. Install Mplfinance and import needed libraries


```python
!pip install mplfinance
!pip install yfinance
import mplfinance as mpf
import pandas as pd
import yfinance as yf
```

# 2. Retrieve stock data for our use case


```python
ticker = "nvda"
nvda = yf.download(ticker, start="2022-01-01", end="2022-06-30")
nvda.head()
```

# 3. Plot the candlesticks and OHLC charts


## A. Candlestick Chart


```python
mpf.plot(nvda['2022-01-01':], type='candle', style = 'classic', title = 'Nvidia Stock Price',tight_layout = True)
```

### Adding Volume Bars and Moving Averages


```python
mpf.plot(nvda['2022-05-01':], type='candle', title = 'Nvidia Stock Price',tight_layout = True, volume = True)
```

```python
mpf.plot(nvda['2022-05-01':], type='candle', title = 'Nvidia Stock Price',tight_layout = True, volume = True, mav = (8))
```

## B. OHLC chart:


```python
mpf.plot(nvda['2022-06-01':], style = 'yahoo', title='Nvidia Stock Price', tight_layout = True )
```

```python
high_low_lines = mpf.make_addplot(nvda.loc['2022-05-01':,['High','Low']])
mpf.plot(nvda["2022-05-01":], addplot=high_low_lines, title = 'Nvidia High and Low Trend', style = 'yahoo', tight_layout = True)
```
