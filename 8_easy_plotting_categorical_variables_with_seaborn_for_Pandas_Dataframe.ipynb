```python
import seaborn as sns 
import matplotlib.pyplot as plt 
from numpy import median 

df_tips = sns.load_dataset('tips') 
df_tips = df_tips.rename( 
                  columns={'total_bill':'Total bill', 
                           'time':'Time', 
                           'sex':'Sex', 
                           'day':'Day'}) 
df_titanic = sns.load_dataset('titanic') 
df_titanic = df_titanic.rename( 
                  columns={'sex':'Sex', 
                           'class':'Class', 
                           'alive':'Alive'})
```

```python
# Set figure size 
plt.figure(figsize=(8,6)) 
# Set theme 
sns.set_theme(style="whitegrid") 
# Plot pointplot 
ax = sns.pointplot(          
                   x="Time",           
                   y="Total bill",           
                   data=df_tips,           
                   estimator=median,           
                   ci='sd',           
                   color="#bb3f3f",           
                   capsize=.1) 
# Set title 
plt.title("Poit plot example with tips dataset") 
# Show figure 
plt.show()
```

```python
# Plot pointplot 
ax = sns.barplot(
                 x="Day",                   
                 y="Total bill",                   
                 data=df_tips,                   
                 estimator=median,                  
                 ci='sd',                   
                 capsize=.07) 
# Set title 
plt.title("Barplot example with tips dataset") 
```

```python
# Plot countplot
ax = sns.countplot(
                   x="Class", 
                   data=df_titanic, 
                   hue='Alive') 
# Set title
plt.title("Countplot example with titanic dataset") 
```

```python
# Plot pointplot 
ax = sns.boxplot(
                 x="Day", 
                 y="Total bill", 
                 data=df_tips) 
# Set title 
plt.title("Boxplot example with tips dataset") 
```

```python
# Plot pointplot 
ax = sns.violinplot(
                    x="Day", 
                    y="Total bill", 
                    data=df_tips) 
# Set title 
plt.title("Violinplot example with tips dataset") 
```

```python
# Plot boxenplot 
ax = sns.boxenplot(
                   x="Day", 
                   y="Total bill", 
                   data=df_tips) 
# Set title 
plt.title("Boxenplot example with tips dataset") 
```

```python
# Plot pointplot 
ax = sns.stripplot(
                   x="Day", 
                   y="Total bill", 
                   data=df_tips) 
# Set title 
plt.title("Stripplot example with tips dataset") 
```

```python
# Plot pointplot 
ax = sns.swarmplot(
                   x="Day", 
                   y="Total bill", 
                   data=df_tips) 
# Set title 
plt.title("Swarmplot example with tips dataset") 
```
