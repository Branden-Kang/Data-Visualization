# A, B

```
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline

figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))

# A
v1 = venn2(subsets=(3, 3, 1), ax=ax1)
c1 = venn2_circles(subsets=(3, 3, 1), ax=ax1)

for area in ['01', '10', '11']:
    color = 'skyblue' if area != '01' else 'white'
    v1.get_patch_by_id(area).set_color(color)
    v1.get_patch_by_id(area).set_alpha(1)
    # If we need to remove numbers, we can use two codes below:
    txt = v1.get_label_by_id(area)
    if txt: txt.set_text('')

ax1.set_axis_on()
ax1.set_facecolor('white')
ax1.set_title('A', fontsize=20)
ymin, ymax = ax1.get_ylim()
ax1.set_ylim(ymin - 0.1, ymax)

# B
v2 = venn2(subsets=(3, 3, 1), ax=ax2)
c2 = venn2_circles(subsets=(3, 3, 1), ax=ax2)

for area in ['01', '10', '11']:
    color = 'skyblue' if area != '10' else 'white'
    v2.get_patch_by_id(area).set_color(color)
    v2.get_patch_by_id(area).set_alpha(1)
    # If we need to remove numbers, we can use two codes below:
    txt = v2.get_label_by_id(area)
    if txt: txt.set_text('')

ax2.set_axis_on()
ax2.set_facecolor('white')
ax2.set_title('B', fontsize=20)
ymin, ymax = ax2.get_ylim()
ax2.set_ylim(ymin - 0.1, ymax)

plt.show()
```

# Complement

```
import pylab as plt
from matplotlib_venn import venn3, venn3_circles
%matplotlib inline

figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))

v1 = venn3(subsets=(1,1,0,1,0,0,0),set_labels = ('','',''), ax=ax1)
c1 = venn3_circles(subsets=(1,1,0,1,0,0,0), ax=ax1)
c1[0].set_lw(0)
c1[2].set_lw(0)

for area in ['001', '100']:
    v1.get_patch_by_id(area).set_color('white')
    txt = v1.get_label_by_id(area)
    if txt: txt.set_text('')
        
v1.get_patch_by_id('010').set_color('skyblue')
v1.get_patch_by_id('010').set_alpha(1)
v1.get_label_by_id('010').set_text('A')
v1.get_label_by_id('010').set_fontsize(20)
v1.get_label_by_id('001').set_text("A'")
v1.get_label_by_id('001').set_fontsize(20)
v1.get_label_by_id('001').set_color('r')

ax1.set_axis_on()
ax1.set_facecolor('white')
ax1.text(-1, 0.2, r'U', fontsize=15)
ax1.set_title("A", fontsize=20)

# Complement
v2 = venn3(subsets=(1,1,0,1,0,0,0),set_labels = ('','',''), ax=ax2)
c2 = venn3_circles(subsets=(1,1,0,1,0,0,0), ax=ax2)
c2[0].set_lw(0)
c2[2].set_lw(0)

for area in ['001', '100']:
    v2.get_patch_by_id(area).set_color('skyblue')
    txt = v2.get_label_by_id(area)
    if txt: txt.set_text('')
        
v2.get_patch_by_id('010').set_color('white')
v2.get_patch_by_id('010').set_alpha(1)
v2.get_label_by_id('010').set_text('A')
v2.get_label_by_id('010').set_fontsize(20)
v2.get_label_by_id('001').set_text("A'")
v2.get_label_by_id('001').set_fontsize(20)
v2.get_label_by_id('001').set_color('r')
v2.get_label_by_id('001').set_rotation(60)

ax2.set_axis_on()
ax2.set_facecolor('skyblue')
ax2.text(-1, 0.2, r'U', fontsize=15)
ax2.set_title("A'", fontsize=20)

plt.show()
```
