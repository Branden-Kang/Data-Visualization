```python
import plotly.express as px
df = px.data.gapminder()
df = df[(df.year == 2002) | (df.year == 2007)]
df.head()
data = df["gdpPercap"].values.tolist()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (12, 12))
plt.violinplot(data, showmedians=True)
plt.show()
```

```python
import seaborn as sns
fig = plt.figure(figsize = (12, 12))
sns.violinplot(x="continent",y="gdpPercap",hue="year",data=df)
```

```python
import plotly.express as px
fig = px.violin(df, y="gdpPercap", x="continent", color="year", box=True, points="all",
          hover_data=df.columns)
fig.show()
```
