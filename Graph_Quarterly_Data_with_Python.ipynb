```python
import matplotlib.pyplot as plt

sales = {2015: [36, 53, 40, 63],
         2016: [46, 93, 99, 70],
         2017: [52, 102, 100, 63],
         2018: [52, 90, 80, 73],
         2019: [75, 120, 150, 120],
         2020: [102, 153, 198, 163],
         2021: [105, 130, 140, 133],
         2022: [120, 202, 180, 153]}

for year, data in sales.items():
    x_coordinates = [year + 0.2 * q for q in [1,2,3,4]]
    plt.bar(x_coordinates,data,width=0.2,edgecolor='black')

years = sales.keys()
plt.xticks([year+0.5 for year in years], years)

plt.title('Sales per Quarter')
plt.ylabel('Sales')
plt.xlabel('Years')

plt.show()
```

```python
net_income = {2018: [9000, 9000, 12000, 15000],
              2019: [-20000, 20000, 15000, 25000],
              2020: [7000, -3000, -9000, -10000],
              2021: [7000, 15000, 15000, 30000],
              2022: [15000, 15000, 25000, 40000]}

for year, data in net_income.items():
    x_coordinates = [year + 0.2 * q for q in [1,2,3,4]]
    plt.bar(x_coordinates,data,width=0.2,edgecolor='black')

years = net_income.keys()
plt.xticks([year+0.5 for year in years], years)

plt.title('Net Income per Quarter')
plt.ylabel('Net Income')
plt.xlabel('Years')

plt.axhline(y=0, color='black', linewidth=0.5)
plt.tight_layout()

plt.show()
```

```python
net_income = {2022: {1:15000, 2:15000, 3:25000},
              2019: {4:25000, 2:20000, 3:15000, 1:-20000},
              2020: {4:-10000, 3:-9000, 2:-3000, 1:7000},
              2018: {4:15000, 2:9000, 3:12000}}


for year, data in net_income.items():
    x_coordinates = [year + 0.2 * q for q in [1,2,3,4]]
    data = [data.get(q,0) for q in [1,2,3,4]]
    plt.bar(x_coordinates,data,width=0.2,edgecolor='black')


years = [y for y in range(min(net_income),max(net_income)+1)]
plt.xticks([year+0.5 for year in years], years)

plt.title('Net Income per Quarter')
plt.ylabel('Net Income')
plt.xlabel('Years')

plt.axhline(y=0, color='black')
```
