```python
# importing the library 
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily
```

```python
# setting the start and end time
start = datetime(2023, 1, 1)
end = datetime(2023, 12, 31)
```

```python
# fetching point data from New Delhi
New_Delhi = Point(28.654667138557485, 77.23127697641567)
```

```python
# using the start and end date for data
data = Daily(New_Delhi, start, end)
data = data.fetch()
```

```python
# visualizing the line chart of temperature 
data.plot(y=['tavg', 'tmin', 'tmax'])
plt.show()
```

```python
# Import Meteostat library
from meteostat import Point

# Create Point New Delhi
New_Delhi = Point(28.654667138557485, 77.23127697641567, 60)
print(New_Delhi)
```

```python
# importing the station method
from meteostat import Stations

# creating an instance of the station class
stations = Stations()

# using the nearby method from the station class
stations = stations.nearby(28.654667138557485, 77.23127697641567)

# fetching one information of weather station
station = stations.fetch(1)

print(station)
```

```python
# Get the number of the weather station in New Delhi
from meteostat import Stations

stations = Stations()

# fetching the number of stations based on the country code and state code
stations = stations.region('IN', 'DL')

print('Stations in New Delhi:', stations.count())
```

```python
from datetime import datetime
from meteostat import Stations

stations = Stations()
stations = stations.nearby(28.654667138557485, 77.23127697641567)

# we can set for daily and hourly basis data
stations = stations.inventory('hourly', datetime(2020, 1, 1))
station = stations.fetch(1)

print(station)
```

```python
from meteostat import Stations, units

stations = Stations()
stations = stations.nearby(28.654667138557485, 77.23127697641567)

# the convert method is used to convert the distance measure to feet
stations = stations.convert({ 'distance': units.feet })

# fetching the nearest 10 weather stations based on the coordinates
stations = stations.fetch(10)

print(stations)
```
