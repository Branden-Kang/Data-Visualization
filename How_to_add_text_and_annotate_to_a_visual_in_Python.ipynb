```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])
plt.text(x=2.2,# x axis of the text
         y=8, # y axis of the text
         s='basic unility of text', #text
         rotation=1,# text rotation
         ha='left',
         va='baseline',
         # the font of text
         fontdict=dict(fontsize=12, color='r',
                       family='monospace',
                       weight='bold',
                      )
        )
```

```python
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])
text = plt.text(x=2.2,# x axis of the text
         y=8, # y axis of the text
         s='basic unility of text', # text
         # font of the text
         fontdict=dict(fontsize=12, color='r',family='monospace',),

         # background box
         bbox={'facecolor': '#74C476', # background color
              'edgecolor':'b',# edge color
               'alpha': 0.5, # transparent
               'pad': 8,# the distance between text and edge
              }

        )

text.set_color('b')# text color
```

```python
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])
plt.text(x=2.2,# x axis of the text
         y=8,# y axis of the text
         s='basic unility of text', #text
         # font of the text
         fontdict=dict(fontsize=12, color='r',family='monospace',),

         #background box
         bbox={'facecolor': '#74C476', # background color
              'edgecolor':'b',# edge color
               'alpha': 0.5, # transparent
               'pad': 0.8,# the distance between text and edge 
               'boxstyle':'roundtooth'
              }

        )
plt.show()
```

```python
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])

plt.annotate('basic unility of annotate', 
             xy=(2, 8),# The posision of arrow sharp

             xytext=(1.0, 8.75),# The position of text

             # Attributes of the arrow
            arrowprops=dict(facecolor='#74C476', 
                            shrink=1,
                            alpha=0.6,
                            width=7,
                            headwidth=40,
                            hatch='--',# Shape of the arrow
                            frac=0.8,
                           ),
            )
```

```python
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])

plt.annotate('basic unility of annotate', 
             xy=(2, 8),
             xytext=(1.0, 8.75),
             arrowprops=dict(facecolor='#74C476', 
                            alpha=0.6,
                            arrowstyle='-|>',# Shape of the arrow
                             color='r'
                           ),
            )
```

```python
plt.figure(figsize=(5,4),dpi=120)
plt.plot([1,2,5],[7,8,9])

plt.annotate('basic unility of annotate', 
             xy=(2, 8),
             xytext=(1.0, 8.75),
             arrowprops=dict(facecolor='#74C476', 
                             alpha=0.6,
                             arrowstyle='-|>',
                             connectionstyle='arc3,rad=0.5',
                             color='r',
                            ),
            )
plt.show()
```

```python
from matplotlib.text import OffsetFrom
fig, (ax1, ax2) = plt.subplots(1, 2,dpi=120)

bbox_args = dict(boxstyle="round", fc="lightgreen")
arrow_args = dict(arrowstyle="->",color='r')

ax1.annotate('Cupid arrow end at ax1', xy=(.5, .75),
                   xytext=(.5, .3), 
                   ha="center", va="center",
                   bbox=bbox_args,
                   arrowprops=dict(connectionstyle="arc3,rad=0.2",
                                   **arrow_args))

ax1.annotate('Stop here', xy=(.5, .7), 
                   xytext=(.5, .8), 
                   ha="center", va="center",
                   bbox=bbox_args,
                   )

ax2.annotate('Cupid arrow start from ax2',
             xy=(0.85, 0.3), xycoords=ax1.transData,# crossing
             xytext=(10, 10),
             textcoords=OffsetFrom(ax2.bbox, (0, 0), "points"),
             ha="left", va="bottom",
             bbox=bbox_args,
             arrowprops=dict(connectionstyle="arc3,rad=0.5",
                             **arrow_args,

                            ))

ax2.set(xlim=[-2, 2], ylim=[-2, 2])
plt.show()
```
