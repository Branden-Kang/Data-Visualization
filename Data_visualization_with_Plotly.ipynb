```python
import numpy as np
import pandas as pd
df = pd.read_csv(
 "https://raw.githubusercontent.com/plotly/datasets/master/volcano_db.csv", encoding="iso-8859-1")
import plotly.express as px
fig = px.histogram(df, x="Country")
fig.show()
```

```python
fig = px.histogram(df, x="Country").update_xaxes(categoryorder="total descending")
fig.show()

```

```python
import plotly.graph_objects as go

labels=df.Status.unique()
values=df['Status'].value_counts()
fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

```python
fig = px.histogram(df, x="Country",color="Status").update_xaxes(categoryorder="total descending")
fig.show()
```

```python
import plotly.express as px
fig = px.bar(df, x='Volcano Name', y='Elev', color='Type', height=400)
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=go.Scattergeo(
        lon = df['Longitude'],
        lat = df['Latitude'],
        mode = 'markers'
        ))

fig.update_layout(
        title = 'Volcanos',
        geo_scope='world',
    )
fig.show()
```

```python
fig = go.Figure(data=go.Scattergeo(
        lon = df['Longitude'],
        lat = df['Latitude'],
        mode = 'markers',
        showlegend=False,
        marker=dict(color="crimson", size=4, opacity=0.8))
        )

fig.update_geos(
    projection_type="orthographic",
    landcolor="white",
    oceancolor="MidnightBlue",
    showocean=True,
    lakecolor="LightBlue"
)
fig.show()
```

```python
df_v = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/volcano.csv")
```

```python
fig = go.Figure(data=[go.Surface(z=df_v.values)])

fig.update_layout(title='Volcano', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
