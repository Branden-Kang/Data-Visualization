```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

# Scatter Plot

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, c ="blue")

# To show the plot
plt.show()
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 4]

plt.bar(categories, values, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

# Histogram

```python
import numpy as np

data = np.random.randn(1000)  # Example data
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

# Box Plot

```python
import numpy as np

# Generating example data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Creating a box plot
plt.boxplot(data, labels=['Box 1', 'Box 2', 'Box 3'], vert=True, patch_artist=True)

# Adding labels and title
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

# Pie Chart

```python
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

# Area Plots

```python
import numpy as np

# Generating example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating an area plot
plt.fill_between(x, y1, y2, color='green', alpha=0.6, label='Area Plot')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('sin - cos Area Plot')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x_values = [1, 2, 3, 4, 5]
y_values = [10, 8, 12, 15, 11]

# Create a line plot
sns.lineplot(x=x_values, y=y_values, color='green', label='Line Plot')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Seaborn Line Plot Example')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn


seaborn.set(style='whitegrid')
fmri = seaborn.load_dataset("fmri")

seaborn.scatterplot(x="timepoint",
     y="signal",
     data=fmri)
```

```python
# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

# Create a bar plot
sns.barplot(x=categories, y=values, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Seaborn Bar Plot Example')

# Show the plot
plt.show()
```

```python
# Example data
data = sns.load_dataset('tips')['total_bill']

# Create a histogram
sns.histplot(data, kde=True, color='salmon', bins=20)

# Add labels and title
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Seaborn Histogram Example')

# Show the plot
plt.show()
```

```python
# Example data
data = sns.load_dataset('tips')

# Create a box plot
sns.boxplot(x='day', y='total_bill', data=data, palette='viridis')

# Add labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.title('Seaborn Box Plot Example')

# Show the plot
plt.show()
```
