```python
# importing required package 
import plotly.express as px
```

# Method 1: From GeoJSON


```python
# reading elections data using plotly express 
df = px.data.election()
df.head()
```

```python
# reading the election geojson file having the coordinates of district of Canada. 
geojson = px.data.election_geojson()
geojson 
```

```python
# creating plot using choropleth function 
fig = px.choropleth(df, geojson = geojson, locations= 'district', featureidkey= "properties.district",
                    projection= "mercator", color = 'Bergeron')

fig.update_geos(fitbounds="locations", visible = False)
```

# Method 2: Using Iso-alpha 3 codes


```python
# reading the gapminder data that contains the data around social and economic conditions of different countries across years.
df = px.data.gapminder()
df.head()
```

```python
# creating map using iso-alpha 3 codes 
fig = px.choropleth(df, locations="iso_alpha", color = "lifeExp", 
                    hover_name= "country",
                    animation_frame = 'year',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig
```
