```python
import plotly.express as px
df = px.data.gapminder().query("year==2007 or year==1997")
```

```python
fig = px.sunburst(df[(df.continent == 'Americas') & (df.year==2007)], path=['continent','country'], values='lifeExp',color='lifeExp')
fig.update_layout(height=600, margin_r=400)
fig.show()
```

```python
fig = px.treemap(df[(df.continent == 'Americas') & (df.year==2007)], path=['continent','country'], values='pop', color='lifeExp')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
fig = px.sunburst(df[df.year==2007], path=['year','continent','country'], values='gdpPercap',color='gdpPercap',color_continuous_scale='ice')
fig.update_layout(height=600, margin_r=400)
fig.show()
```

```python
fig = px.treemap(df[df.year==2007], path=['year','continent','country'], values='gdpPercap',color='gdpPercap',color_continuous_scale='hot')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
import numpy as np
fig = px.sunburst(df[df.continent == 'Americas'], path=['year','continent','country'], values='lifeExp',color='lifeExp', color_continuous_scale='ice', color_continuous_midpoint=np.average(df['lifeExp']))
fig.update_layout(height=600, margin_r=400)
fig.show()
```

```python
fig = px.treemap(df, path=[px.Constant("world"), 'year','continent', 'country'], values='lifeExp', color='lifeExp',color_continuous_scale='hot')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
fig = px.sunburst(df, path=['year','continent'], values='pop', labels='pop', color='pop',color_continuous_scale='ice')
fig.update_traces(textinfo="label+percent root")
fig.update_layout(height=600, margin_r=400)
fig.show()
```
