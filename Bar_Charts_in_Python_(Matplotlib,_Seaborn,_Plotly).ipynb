```python
import matplotlib.pyplot as plt

#fake data
data = {'THY':201, 'RYAN':55, 'EMIRATES':300, 'QATAR':400, 'LUFTHANSA':198}
lines = list(data.keys())
sells = list(data.values())

#figure
fig = plt.figure(figsize = (12, 5))
plt.bar(lines, sells, color ='blue',
        width = 0.5)
plt.xlabel("Airlines")
plt.ylabel("Ticket Sold")
plt.title("Fake Data of Airlines Sales")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = {'THY':201, 'RYAN':55, 'EMIRATES':300, 'QATAR':400, 'LUFTHANSA':198}
lines = list(data.keys())
sells = list(data.values())
ax = sns.barplot(x = lines,y = sells,)
ax.set(xlabel='Airlines', ylabel='Tickets Sold', title="Fake Data of Airlines")
plt.show()
```

```python
data = sns.load_dataset('titanic')
 
sns.barplot(x = 'class', y = 'fare',
            hue = 'sex', data = data, 
            palette='bright')
 
ax.set(title="Genders of Fare's in Titanic Dataset")
plt.show()
```

```python
import plotly.express as px
data = px.data.gapminder().query("country == 'Germany'")
fig = px.bar(data, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], 
             color='lifeExp',)
fig.show()
```
