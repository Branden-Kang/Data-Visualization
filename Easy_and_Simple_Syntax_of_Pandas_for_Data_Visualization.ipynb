```python
import pandas as pd

url = 'https://raw.githubusercontent.com/shoukewei/data/main/data-pydm/gdp_top_six_economies.csv'
df = pd.read_csv(url,index_col='Year') 
df.head()
```

```python
df.columns
```

```python
df.plot(kind='line')
```

```python
df.plot.line()
```

```python
df.plot.line(y=['United States','China','Japan','Germany'], 
             title='GDP of world top 4 economies', 
             ylabel='GDP (current US$)',
             figsize=(8,4))
```

```python
axes = df.plot.line(subplots=True, 
                    y=['United States','China','Japan','Germany'], 
                    title='GDP of world top 4 economies', 
                    ylabel='GDP (current US$)',
                    figsize=(10, 8))
```

```python
axes = df.plot.line(subplots=True, 
                    y=['United States','China','Japan','Germany'], 
                    title='GDP of world top 4 economies', 
                    ylabel='GDP (current US$)',
                    figsize=(10, 8),
                    layout=(2,2))
```

```python
gdp_mean = df.agg('mean')
gdp_mean
```

```python
gdp_mean.plot.bar(figsize=(8,4),
                  xlabel='Multi-anual average GDP',
                  ylabel='GDP (current US$)')
```

```python
df_3y = df[-3:]
df_3y
```

```python
df_3y.plot.bar(figsize=(8,4),
               xlabel='Recent 3 years',
               ylabel='GDP (current US$)')
```

```python
df_3y.plot.barh(figsize=(8,5),
                xlabel='Recent 3 years',
                ylabel='GDP (current US$)'
                )
```

```python
df_3y.plot.bar(stacked=True,
               figsize=(8,4),
               xlabel='Recent 3 years',
               ylabel='GDP (current US$)'
               )
```

```python
df['United States'].plot.hist(
    bins=25,
    figsize=(8,4))
```

```python
df[['United States', 'China']].plot.hist(
    bins=25,
    alpha=0.6, 
    figsize=(8,4))
```

```python
df[['United States', 'China']].plot.hist(
    bins=25, 
    alpha=0.6,                              
    stacked=True,
    figsize=(8,4))
```

```python
df.plot.box(figsize=(8,4),
            rot=45)
```

```python
color = {
    "boxes": "DarkGreen",
    "whiskers": "DarkOrange",
    "medians": "DarkBlue",
    "caps": "Gray",
}

df.plot.box(color=color, 
            sym="r+",
            vert=False,
            figsize=(8,6))
```

```python
df.plot.area(figsize=(8,4),
            ylabel='GDP (current US$)')
```

```python
df_3y.index=['2018', '2019', '2020']
pie = df_3y.plot.pie(y='United States',  
                     legend=True,
                     autopct='%.2f',
                     rotatelabels=45)
pie.legend(loc='upper left')
```

```python
df.plot.scatter(x='China',
                y='United States',  
                figsize=(8,4),  
                color='Green')
```

```python
df.plot.hexbin(x='China', 
               y='United States', 
               figsize=(8,4))
```

```python
df.plot.hexbin(x='China', 
               y='United States', 
               gridsize=10, 
               figsize=(8,4))
```

```python
df.plot.kde(figsize=(8,4))
```

```python
df.plot.kde(figsize=(8,4), 
            bw_method=0.1)
```

```python
df.plot.kde(figsize=(8,4), 
            bw_method=1)
```
