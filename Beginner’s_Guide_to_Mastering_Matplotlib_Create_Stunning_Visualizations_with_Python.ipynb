```python
import matplotlib.pyplot as plt

# Get ready to jazz up your plot styles, colors, and markers!
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linestyle='--', marker='o', color='blue')
plt.scatter(x, y, c='red', marker='s')
plt.bar(x, y, color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Boost clarity with titles, labels, and legends!
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, linestyle='--', marker='o', color='blue')
plt.title('A Tale of Customization')
plt.xlabel('X Marks the Spot')
plt.ylabel('Y Represents Why')
plt.legend(['Line Plot'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating the line plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sin(x)')
plt.title('Line Plot of Sin(x)')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.legend()
plt.show()
```

```python
# Bar chart example
categories = ['Category 1', 'Category 2', 'Category 3']
values = [10, 15, 7]

plt.figure(figsize=(7, 4))
plt.bar(categories, values, color='blue')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Pie chart example
labels = ['Part A', 'Part B', 'Part C', 'Part D']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
# Scatter plot example
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
# Histogram example
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.5, color='g', density=True)
plt.title('Histogram Example')
plt.xlabel('Data Points')
plt.ylabel('Probability')
plt.show()
```

```python
# Box plot example
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)
ax.plot(x, y, z)
ax.set_title('3D Line Plot Example')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
