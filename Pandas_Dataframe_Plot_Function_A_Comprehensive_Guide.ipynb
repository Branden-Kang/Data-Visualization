# Line Plots

```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 

df = pd.DataFrame() 
df['date'] = pd.date_range('1/1/2022', periods=365) 
df['stock_price'] = [i*2 for i in range(365)] 
df.plot(x='date', y='stock_price')
```

# Bar Plots

```python
df = pd.DataFrame() 
df['movie'] = ['Movie A', 'Movie B', 'Movie C'] 
df['ticket_sales'] = [100, 200, 300] 
df.plot(kind='bar', x='movie', y='ticket_sales') 
```

# Advanced Options for Customizing Your Plot

```python
df = pd.DataFrame() 
df['date'] = pd.date_range('1/1/2022', periods=365) 
df['stock_price'] = [i*2 for i in range(365)] 
df.plot(x='date', y='stock_price', linestyle='--')
```

```python
df.plot(x='date', 
        y='stock_price', 
        linestyle='--', 
        color='green', 
        title='Awesome Plot')
```
