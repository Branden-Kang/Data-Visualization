```python
# create dataset 1
x1 = [*range(0,10+1)]
y1 = [-i*(i-14) for i in x1]
print(x1, y1)
```

```python
# create dataset 2
x2 = [*range(0,5+1)]
y2 = [3*i for i in x2]
print(x2, y2)
```

```python
# create dataset 3
x3 = [*range(5,10+1)]
y3 = [15+(i-5)**2 for i in x3]
print(x3, y3)
```

```python
import matplotlib.pyplot as plt
plt.plot(x1, y1, x2, y2, x3, y3)
plt.show()
```

```python
# Axes labels and grid
import matplotlib.pyplot as plt
plt.plot(x1, y1, x2, y2, x3, y3)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# Legend
import matplotlib.pyplot as plt
plt.plot(x1, y1, label='dataset 1')
plt.plot(x2, y2, label='dataset 2')
plt.plot(x3, y3, label='dataset 3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Line color
plt.plot(x1, y1, 'r', label='red')
plt.plot(x2, y2, 'g', label='green')
plt.plot(x3, y3, 'b', label='blue')
```

```python
plt.plot(x1, y1, 'b', label='blue')
plt.plot(x2, y2, 'b')
plt.plot(x3, y3, 'b')
```

```python
# Linestyle
plt.plot(x1, y1, '-', label='solid')
plt.plot(x2, y2, '--', label='dashed')
plt.plot(x3, y3, ':', label='dotted')
```

```python
plt.plot(x1, y1, 'b-.', label='blue')
plt.plot(x2, y2, 'b--')
plt.plot(x3, y3, 'b-')
```

```python
# Marker
plt.plot(x1, y1, 's', label='A')
plt.plot(x2, y2, '^', label='B')
plt.plot(x3, y3, 'o', label='C')
```

```python
# Color, linestyle, marker combination?
plt.plot(x1, y1, 'sr:', label='A')
plt.plot(x2, y2, '-.g^', label='B')
plt.plot(x3, y3, 'o-b', label='C')
```

```python
# Axes limit
import matplotlib.pyplot as plt
plt.plot(x1, y1, 'sr:', label='A')
plt.plot(x2, y2, '-.g^', label='B')
plt.plot(x3, y3, 'o-b', label='C')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.xlim([-2, 12])
plt.ylim([-10, 60])
plt.show()
```

```python
# Ticks for x and y axes
import matplotlib.pyplot as plt
plt.plot(x1, y1, 'sr:', label='A')
plt.plot(x2, y2, '-.g^', label='B')
plt.plot(x3, y3, 'o-b', label='C')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.xlim([-2, 12])
plt.ylim([-10, 60])

plt.xticks([i for i in range(-2, 12+1)])
plt.yticks([i for i in range(-10, 60+1, 5)])
plt.show()
```

```python
# Subplo
import matplotlib.pyplot as plt

plt.subplot(3, 3, 3)
plt.plot(x1, y1, 'sr:')
plt.title("Fig A")

plt.subplot(3, 3, 5)
plt.plot(x2, y2, '-.g^')
plt.title("Fig B")

plt.subplot(3, 3, 7)
plt.plot(x3, y3, 'o-b')
plt.title("Fig C")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(3, 3, 3)
plt.plot(x1, y1, 'sr:')
plt.title("Fig A")
plt.xlabel('x1')
plt.ylabel('y1')
plt.grid()

plt.subplot(3, 3, 5)
plt.plot(x2, y2, '-.g^')
plt.title("Fig B")
plt.xlabel('x2')
plt.ylabel('y2')
plt.grid()

plt.subplot(3, 3, 7)
plt.plot(x3, y3, 'o-b')
plt.title("Fig C")
plt.xlabel('x3')
plt.ylabel('y3')
plt.grid()

plt.show()
```

```python
# Annotation
import matplotlib.pyplot as plt

plt.plot(x1, y1, 'sr:')
plt.title("Fig A")
plt.xlabel('x1')
plt.ylabel('y1')
plt.text(0.3, -0.1, 'P')
plt.text(9.9, 42, 'Q')
plt.grid()
```

```python
# A wrap-up of examples
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'sr:', label='A')
plt.legend()
plt.ylabel('y')
plt.title('quadratic')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, '-.g^', label='B')
plt.legend()
plt.title('linier')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'o-b', label='C')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x1, y1, 'sr:')
plt.plot(x2, y2, '-.g^')
plt.plot(x3, y3, 'o-b')
plt.xlabel('x')
plt.xticks([0, 3, 4, 6, 8, 9, 10])
plt.text(5, 30, "area")

plt.show()
```
