```python
import matplotlib.pyplot as plt

fixed_costs = 1500
variable_costs = 0.50
sales_price = 1.5

bep = fixed_costs / (sales_price-variable_costs)

plt.plot([0, 2 * bep],[fixed_costs, fixed_costs + variable_costs * 2 * bep],
         color='red',label='Costs')
plt.plot([0, 2 * bep],[0, sales_price * 2 * bep],
         color='green',label='Revenue')


plt.ylim(bottom=0)
plt.xlim(left=0)
plt.xlim(right=2*bep)

plt.legend()
plt.title('Break-Even Analysis Ice Cream Business')
plt.ylabel('Dollars')
plt.xlabel('Sales')

plt.show()
```
