```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
sales = pd.read_csv("https://raw.githubusercontent.com/Branden-Kang/Data-Visualization/master/Data/supermarket_sales.csv")
sales.head()
```

```python
sales.isnull().sum()
```

```python
fig = px.histogram(sales, x="Product line", y="Total", color='Payment', barmode= 'group')
fig.show() #to display plot
```

```python
#Product sales by product line and by payment method
fig = px.histogram(sales, x="Product line", y="Total", color='Payment', barmode="group")

fig.update_xaxes(title_text="Product Line")
fig.update_yaxes(title_text="Total")
fig.update(layout_title_text="Product sales by Product Line and Payment Method",)

fig.show()
```

```python
piec = px.pie(sales, names = "Product line", values ="Unit price", color_discrete_sequence=px.colors.sequential.Viridis)
piec.update_traces(marker=dict(line=dict(color='#000000', width=2)))
piec.show()
```

```python
#Donut Chart
#Unit Price by Product Line

don = px.pie(sales, names = "Product line", values ="Unit price", hole=0.5, color_discrete_sequence=px.colors.sequential.Viridis)
don.update_traces(marker=dict(line=dict(color='#000000', width=2)))
don.update_layout(title_text='PRODUCT LINE BY UNIT PRICE', legend_title_text='PRODUCT LINE')

don.show()
```

```python
fig = px.box(sales, x="Branch", y ="gross income", color = "Branch", title = "Gross income made by branches")

fig.show()
```
